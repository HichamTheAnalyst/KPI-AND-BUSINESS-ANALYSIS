# KEP PERFORMANCE INDICATOR (KPI)

1. **Taux de Croissance du Chiffre d'Affaires** :
   \[
   \text{Taux de Croissance du CA} = \left( \frac{\text{Chiffre d'Affaires de la Période Actuelle} - \text{Chiffre d'Affaires de la Période Précédente}}{\text{Chiffre d'Affaires de la Période Précédente}} \right) \times 100\%
   \]
   - Calculez le chiffre d'affaires pour la période actuelle et celui de la période précédente (par exemple, année en cours vs année précédente).
   - Soustrayez le chiffre d'affaires de la période précédente du chiffre d'affaires de la période actuelle.
   - Divisez le résultat par le chiffre d'affaires de la période précédente et multipliez par 100 pour obtenir le taux de croissance en pourcentage.

2. **Marge Brute** :
   \[
   \text{Marge Brute} = \left( \frac{\text{Chiffre d'Affaires} - \text{Coût des Biens Vendus}}{\text{Chiffre d'Affaires}} \right) \times 100\%
   \]
   - Calculez le chiffre d'affaires et le coût des biens vendus (COGS).
   - Soustrayez le COGS du chiffre d'affaires pour obtenir le bénéfice brut.
   - Divisez le bénéfice brut par le chiffre d'affaires et multipliez par 100 pour obtenir la marge brute en pourcentage.

3. **Marge Nette** :
   \[
   \text{Marge Nette} = \left( \frac{\text{Bénéfice Nette}}{\text{Chiffre d'Affaires}} \right) \times 100\%
   \]
   - Calculez le bénéfice net (chiffre d'affaires moins toutes les dépenses).
   - Divisez le bénéfice net par le chiffre d'affaires et multipliez par 100 pour obtenir la marge nette en pourcentage.

4. **Ratio de Liquidité Courante** :
   \[
   \text{Ratio de Liquidité Courante} = \frac{\text{Actifs Courants}}{\text{Passifs Courants}}
   \]
   - Calculez le total des actifs courants (actifs devant être convertis en espèces dans l'année) et le total des passifs courants (obligations dues dans l'année).
   - Divisez le total des actifs courants par le total des passifs courants pour obtenir le ratio de liquidité courante.

5. **Ratio de Liquidité Rapide (Ratio de Test Acide)** :
   \[
   \text{Ratio de Liquidité Rapide} = \frac{\text{Actifs Courants - Stocks}}{\text{Passifs Courants}}
   \]
   - Calculez le total des actifs courants moins les stocks (car les stocks peuvent ne pas être facilement convertibles en espèces).
   - Divisez le résultat par le total des passifs courants pour obtenir le ratio de liquidité rapide.

6. **Cycle de Conversion de Trésorerie (CCC)** :
   \[
   \text{CCC} = \text{Délai Moyen de Rotation des Stocks} + \text{Délai Moyen de Règlement des Créances} - \text{Délai Moyen de Paiement des Dettes}
   \]
   - Calculez le nombre moyen de jours nécessaires pour vendre les stocks (Délai Moyen de Rotation des Stocks), collecter les créances (Délai Moyen de Règlement des Créances) et payer les dettes (Délai Moyen de Paiement des Dettes).
   - Additionnez le Délai Moyen de Rotation des Stocks et le Délai Moyen de Règlement des Créances, puis soustrayez le Délai Moyen de Paiement des Dettes pour obtenir le CCC.

Ces formules fournissent une base pour calculer chacun des KPI mentionnés, vous permettant d'évaluer efficacement la performance financière de votre entreprise.

### Revenue Growth Rate:

#### Measures the percentage increase in revenue over a specific period.

```python
previous_revenue = 10000
current_revenue = 12000
growth_rate = ((current_revenue - previous_revenue) / abs(previous_revenue)) * 100 if previous_revenue != 0 else 0
print("Revenue Growth Rate:", growth_rate)
```

### Customer Acquisition Cost (CAC):

#### Measures the cost to acquire a new customer.

```python
total_sales_marketing_cost = 10000
new_customers_acquired = 100
cac = total_sales_marketing_cost / new_customers_acquired if new_customers_acquired != 0 else 0
print("Customer Acquisition Cost (CAC):", cac)
```

### Customer Lifetime Value (CLV):

#### Measures the total revenue a customer is expected to generate throughout their relationship with the company.

```python
average_revenue_per_customer = 1000
churn_rate = 0.05
clv = average_revenue_per_customer / churn_rate if churn_rate != 0 else 0
print("Customer Lifetime Value (CLV):", clv)
```

### Churn Rate:

#### Measures the percentage of customers who stop using a product or service within a given time period.

```python
customers_lost = 50
total_customers_start = 1000
churn_rate = (customers_lost / total_customers_start) * 100 if total_customers_start != 0 else 0
print("Churn Rate:", churn_rate)
```

### Customer Satisfaction (CSAT):

#### Measures the level of satisfaction customers have with a product or service.

```python
total_satisfaction_score = 350   # Total score from all responses
total_responses = 100            # Total number of responses
csat = total_satisfaction_score / total_responses if total_responses != 0 else 0
print("Customer Satisfaction (CSAT):", csat)
```

### Net Promoter Score (NPS):

#### Measures the likelihood of customers to recommend a product or service to others.

```python
promoters = 200
detractors = 50
total_responses = 500
nps = ((promoters - detractors) / total_responses) * 100 if total_responses != 0 else 0
print("Net Promoter Score (NPS):", nps)
```

### Average Revenue Per User (ARPU):

#### Measures the average revenue generated per user or customer.

```python
total_revenue = 10000
total_users = 500
arpu = total_revenue / total_users if total_users != 0 else 0
print("Average Revenue Per User (ARPU):", arpu)
```

### Conversion Rate:

#### Measures the percentage of users or visitors who take a desired action, such as making a purchase or signing up for a service.

```python
conversions = 100
total_visitors = 1000
conversion_rate = (conversions / total_visitors) * 100 if total_visitors != 0 else 0
print("Conversion Rate:", conversion_rate)
```

### Gross Margin:

#### Measures the percentage of revenue that exceeds the cost of goods sold.

```python
revenue = 50000
cogs = 30000
gross_margin = ((revenue - cogs) / revenue) * 100 if revenue != 0 else 0
print("Gross Margin:", gross_margin)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Data for KPIs
kpis = ['Revenue Growth Rate', 'Customer Acquisition Cost', 'Customer Lifetime Value', 
        'Churn Rate', 'Customer Satisfaction', 'Net Promoter Score', 
        'Average Revenue Per User', 'Conversion Rate', 'Gross Margin']
values = [20, 30, 40, 15, 80, 60, 25, 12, 50]  # Example values for visualization

# Create a DataFrame for visualization
import pandas as pd
df = pd.DataFrame({'KPI': kpis, 'Value': values})

# Set Seaborn style
sns.set_style("whitegrid")

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x='Value', y='KPI', data=df, palette='viridis')
plt.xlabel('KPI Values')
plt.ylabel('')
plt.title('Key Performance Indicators')
plt.show()
```

### CASH FLOW 

```python
def calculate_operating_cash_flow(net_income, depreciation, changes_in_working_capital):
    return net_income + depreciation + changes_in_working_capital

# Example usage:
net_income = 50000
depreciation = 10000
changes_in_working_capital = 2000  # Positive value means cash inflow, negative means cash outflow
ocf = calculate_operating_cash_flow(net_income, depreciation, changes_in_working_capital)
print("Operating Cash Flow (OCF):", ocf)
```

```python
import matplotlib.pyplot as plt

def visualize_cash_flow(net_income, depreciation, changes_in_working_capital):
    labels = ['Net Income', 'Depreciation', 'Changes in Working Capital', 'Operating Cash Flow']
    values = [net_income, depreciation, changes_in_working_capital, net_income + depreciation + changes_in_working_capital]

    # Create a waterfall chart
    plt.figure(figsize=(10, 6))
    plt.bar(labels, values, color=['blue', 'green', 'orange', 'purple'])
    plt.plot(values, marker='o', linestyle='-', color='black')
    plt.title('Operating Cash Flow Waterfall Chart')
    plt.xlabel('Components')
    plt.ylabel('Amount ($)')
    plt.grid(True)
    plt.show()

# Example usage:
net_income = 50000
depreciation = 10000
changes_in_working_capital = 2000
visualize_cash_flow(net_income, depreciation, changes_in_working_capital)
```

```python
def calculer_fonds_de_roulement(actifs_courants, passifs_courants):
    return actifs_courants - passifs_courants

# Exemple d'utilisation :
actifs_courants = 100000
passifs_courants = 75000
fonds_de_roulement = calculer_fonds_de_roulement(actifs_courants, passifs_courants)
print("Fonds de roulement:", fonds_de_roulement)
```

```python
import matplotlib.pyplot as plt

def visualiser_fonds_de_roulement(actifs_courants, passifs_courants):
    labels = ['Actifs Courants', 'Passifs Courants', 'Fonds de Roulement']
    valeurs = [actifs_courants, passifs_courants, actifs_courants - passifs_courants]

    # Créer un graphique en barres pour visualiser le fonds de roulement
    plt.figure(figsize=(10, 6))
    plt.bar(labels, valeurs, color=['blue', 'orange', 'green'])
    plt.title('Visualisation du Fonds de Roulement')
    plt.xlabel('Composants')
    plt.ylabel('Montant ($)')
    plt.grid(True)
    plt.show()

# Exemple d'utilisation :
actifs_courants = 100000
passifs_courants = 75000
visualiser_fonds_de_roulement(actifs_courants, passifs_courants)
```

```python
def calculer_besoin_fonds_roulement(stocks, creances_clients, dettes_fournisseurs):
    return stocks + creances_clients - dettes_fournisseurs

# Exemple d'utilisation :
stocks = 50000
creances_clients = 30000
dettes_fournisseurs = 20000
besoin_fonds_roulement = calculer_besoin_fonds_roulement(stocks, creances_clients, dettes_fournisseurs)
print("Besoin en fonds de roulement:", besoin_fonds_roulement)
```

```python
import matplotlib.pyplot as plt

def visualiser_composants_bfr(stocks, creances_clients, dettes_fournisseurs):
    labels = ['Stocks', 'Créances Clients', 'Dettes Fournisseurs']
    valeurs = [stocks, creances_clients, dettes_fournisseurs]

    # Créer un graphique en barres pour visualiser les composants du BFR
    plt.figure(figsize=(10, 6))
    plt.bar(labels, valeurs, color=['blue', 'green', 'orange'])
    plt.title('Composants du Besoin en Fonds de Roulement')
    plt.xlabel('Composants')
    plt.ylabel('Montant ($)')
    plt.grid(True)
    plt.show()

# Exemple d'utilisation :
stocks = 50000
creances_clients = 30000
dettes_fournisseurs = 20000
visualiser_composants_bfr(stocks, creances_clients, dettes_fournisseurs)
```
